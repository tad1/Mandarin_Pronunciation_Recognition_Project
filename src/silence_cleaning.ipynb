```python
import os
from IPython.display import Audio

def play_wav(rec_path: str):
    # Normalize path to current OS
    wav_path = os.path.join(AUDIO_PATH, *rec_path.split("/"))
    return Audio(filename=wav_path)
```

```python
import torchaudio
import torchaudio.transforms as T
import matplotlib.pyplot as plt
import torch

from audio import load_audio, PROJECT_SAMPLING_RATE
import os

def show_mel(path, sr=22050, n_mels=128):
    full_path = os.path.join(AUDIO_PATH, *path.split("/"))
    waveform = load_audio(full_path, "torchaudio") 
    mel_spectrogram = T.MelSpectrogram(sample_rate=PROJECT_SAMPLING_RATE, n_mels=n_mels)(waveform)
    mel_db = T.AmplitudeToDB()(mel_spectrogram)

    plt.figure(figsize=(8, 3))
    plt.imshow(mel_db.numpy(), aspect="auto", origin="lower")
    plt.title("Mel Spectrogram: " + path)
    plt.ylabel("Mel Bin")
    plt.xlabel("Frame")
    plt.colorbar(label="dB")
    plt.tight_layout()
    plt.show()

```

```python
import torchaudio
import torchaudio.transforms as T
import os
from audio import load_audio, PROJECT_SAMPLING_RATE

def trim_and_save(path: str, db_threshold: float = 60.0):
    full_path = os.path.join(AUDIO_PATH, *path.split("/"))
    waveform = load_audio(full_path, "torchaudio")


    # Trim silence (default: leading + trailing silence below threshold dB)
    trimmed_waveform= torchaudio.functional.vad(
        waveform, sample_rate=PROJECT_SAMPLING_RATE, trigger_level=db_threshold
    )

    # If VAD fails (no voice), fallback to manual trim
    if trimmed_waveform.numel() == 0:
        trimmed_waveform = torchaudio.transforms.Vad(PROJECT_SAMPLING_RATE)(waveform)

    # Save with _t suffix
    new_path = full_path.replace(".wav", "_t.wav")
    torchaudio.save(new_path, trimmed_waveform.reshape((1,-1)), PROJECT_SAMPLING_RATE)
    print(f"Saved trimmed file: {new_path}")
```

```python
def trim_compare(path:str):
    show_mel(path)
    display(play_wav(path))
    trim_and_save(path=path)
    new_path = path.replace(".wav", "_t.wav")
    show_mel(new_path)
    display(play_wav(new_path))
```

```python
for path in filtered.select("rec_path").to_series():
    trim_compare(path)
```
