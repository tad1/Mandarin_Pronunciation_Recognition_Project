```python
import numpy as np
import tensorflow as tf
import keras
from keras import layers
```

```python
WINDOW_SIZE = 512
SLIDE_SIZE = 256
NUM_CLASSES = 5
```

```python
from tf_dataset import get_tf_dataset
dataset = get_tf_dataset(1000)
```

```python
EPOCHS = 40




batched_dataset = dataset.batch(100, drop_remainder=True)
training = batched_dataset.skip(1).take(6)
validation = batched_dataset.skip(7).take(1)
# training = dataset.take(200)

print(training)

```

```python
import time
import keras
model = keras.Sequential([
    # https://stackoverflow.com/questions/62031683/ragged-tensors-as-input-for-lstm
    # layers.Input(shape=(1099,1), batch_size=128),
    layers.LSTM(NUM_CLASSES, return_sequences=True, unroll=True),
    # layers.GRU(NUM_CLASSES, return_sequences=True, unroll=True),
    layers.Dense(NUM_CLASSES, activation='log_softmax'),
    # layers.Dense(NUM_CLASSES, activation='relu'),
])

loss_fn = tf.keras.losses.CTC(reduction='mean')
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
# model.summary()

# model.fit(training, epochs=EPOCHS)

for epoch in range(EPOCHS):
    for val, lab in training:
        val = tf.sparse.from_dense(val)

        with tf.GradientTape() as tape:
            predictions = model(val)
            loss = loss_fn(lab, predictions)
        grads = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        print(f"Epoch {epoch}, Loss: {loss}")
        # validate
    for val, lab in validation:
        predictions = model(val)
        loss = loss_fn(lab, predictions)
        print(f"Validation Loss: {loss}")


        # val = pre_process(validation[0], WINDOW_SIZE)
        # lab = pre_process_labels(validation_labels[0])
        # lab = tf.reshape(lab, (1, -1))
        # predictions = model(val)
        # loss = loss_fn(lab, predictions)
        # print(f"Validation Loss: {loss}")
```
